```python
# Import pandas
import pandas as pd

# Example student dataset (with Aadhar card and Mentor names)
data = {
    'rollno': [101, 102, 103, 104, 105],
    'name': ['John', 'Alice', 'Bob', 'Eve', 'Charlie'],
    'gender': ['Male', 'Female', 'Male', 'Female', 'Male'],
    'state': ['California', 'Texas', 'Delhi', 'Delhi', 'California'],
    'mentor_name': ['Dr. Mamba', 'Ms. Johnson', 'Dr. Mamba', 'Ms. Davis', 'Mr. Kim'],
    'marks': [88, 92, 79, 95, None],  # Including a null value for marks
    'aadhar_card': ['1234-5678-9123', None, '9876-5432-1234', '4567-1234-5678', None]  # Some Aadhar cards are missing
}

# Create a DataFrame from the data
df = pd.DataFrame(data)

# Count the number of rows and columns
rows, cols = df.shape
print(f"Rows: {rows}, Columns: {cols}")

# Count the number of Male and Female students
gender_count = df['gender'].value_counts()
print(f"Gender count:\n{gender_count}")

# Count the students by state
state_count = df['state'].value_counts()
print(f"State-wise count:\n{state_count}")

# Count the number of null values in the DataFrame
null_count = df.isnull().sum()
print(f"Null value count:\n{null_count}")

# Extract specific columns: rollno, name, mentor name
specific_columns = df[['rollno', 'name', 'mentor_name']]
print(f"Extracted columns:\n{specific_columns}")

# List all students where the mentor name is 'Dr. Mamba'
dr_mamba_students = df[df['mentor_name'] == 'Dr. Mamba']
print(f"Students with mentor Dr. Mamba:\n{dr_mamba_students}")

# List all students from Delhi
delhi_students = df[df['state'] == 'Delhi']
print(f"Students from Delhi:\n{delhi_students}")

# List all students whose Aadhar card is not available (null in 'aadhar_card' column)
students_without_aadhar = df[df['aadhar_card'].isnull()]
print(f"Students without Aadhar card:\n{students_without_aadhar}")
```
